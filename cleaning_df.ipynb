```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\User\Desktop\project in data science\ProjectCsv\Not_cleand.csv",dtype={'col':np.float})
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df.rename(columns={'Unnamed: 0.1':'Country'},inplace=True)
df.set_index('Country',inplace=True)
df=df.T
df
```

```python
for i in df.columns:
    df.set_index(i)
df.dropna(axis=0, thresh=170,inplace=True)


df
```

```python
df.dropna(axis=1, thresh=160,inplace=True)
df.drop('911 by Country',axis=1,inplace=True)
df.drop('GNP/GNI by Country',axis=1,inplace=True)
df.drop(df.columns[[7]],axis=1,inplace=True)
df.drop(df.columns[[2]],axis=1,inplace=True)
df
```

```python

df = df.replace(',','', regex=True)
df = df.replace('-',np.nan, regex=True)
df = df.replace('--',np.nan, regex=True)
df= df.astype(str)



for i in df.columns:
    if(  (df[i].iat[0]).isnumeric()==True ):
        df[i]=df[i].astype(float)
        
    else:
            if('%' in (df[i].iat[0]) ):
                df[i]=df[i].str.strip("%")
                df[i]=df[i].str.strip(",")
                df[i]=df[i].astype(float)/100
            elif('$' in (df[i].iat[0]) ):
                for s in range(0,len(df[i].values)):
                    df[i].values[s]=df[i].values[s].strip("$")
                    if ('Bn' in df[i].values[s] ):
                        df[i].values[s]=df[i].values[s].strip("Bn")
                        df[i].values[s]=float(df[i].values[s])*pow(10,9)
                    elif('Tn' in df[i].values[s]  ):
                        df[i].values[s]=df[i].values[s].strip("Tn")
                        df[i].values[s]=float(df[i].values[s])*pow(10,12)
                    elif('Mn' in df[i].values[s]  ):
                        df[i].values[s]=df[i].values[s].strip("Mn")
                        df[i].values[s]=float(df[i].values[s])*pow(10,6)
                df[i]=df[i].astype(float)
            
                 
                
              
                    
df = df.replace('NaN',np.nan, regex=True)
df = df.replace('nan',np.nan, regex=True)
df=df.replace(0.0,np.nan, regex=True)
df=df.replace('0.0',np.nan, regex=True)
df.dropna(axis=1, thresh=120,inplace=True)

df
```

```python

for i in df.columns:
    try:
        if(float(df[i].iat[0])):
            df[i]=df[i].astype(float)
            continue
    except :
        pass
        
    

    
for i in df.columns:
    if(df[i].dtypes!=object):
        df[i].fillna(df[i].mean(), inplace=True)
        

df
```
