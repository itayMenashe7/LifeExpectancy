```python
import requests
from bs4 import BeautifulSoup as bs
import re
import time
from collections import Counter
import pandas as pd
```

```python
dict_countrys={}
page_country='https://worldpopulationreview.com/'
respones=requests.get(page_country)
page=bs(respones.text,'html.parser')
table=page.find_all('tbody')
for row in table[1].find_all('tr'):
    for ppl in row.find_all('td')[1]:
        country_value=ppl.text
    for ppl in row.find_all('td')[3]:
        people_value=float(ppl.replace(',', ''))
    dict_countrys[country_value]=people_value 
dict_countrys
```

```python
page='https://worldpopulationreview.com/countries'
respones=requests.get(page)
page=bs(respones.text,'html.parser')
page_names=[] ## list of possible list name
all_possible_hrefs=[] ## list of possible links in the site
possible_dict={} ## Dict of all countries Subject:link
temp=[]## Tempory array to help remove the duplictes that resulted from scrapping helps order the dict if dict not needed may be removed
for link in page.find_all('section'):
    if((link.find('div',{'class':'actionbar py-1 px-2'})) and not(link.find('table')) and not(link.find('div',{'class':'letter-list'})) 
      and not link.find('ol') ):
        page_name=link.find('div',{'class':'actionbar py-1 px-2'})
        page_names.append(page_name.div.text)
        possible_dict[page_name.div.text]=[]
        for possible_urls in link.find_all('a'):
            all_possible_hrefs.append(possible_urls.get('href'))
            temp.append(possible_urls.get('href'))
    remove_dup=list(set(temp))
    if(temp):
        possible_dict[page_name.div.text].extend(remove_dup)
    
        
unique = list(set(all_possible_hrefs))
print(unique)

```

```python
def func_find_number(text):
    if (('100K' in text) or ('100,000' in text)): 
        return 100000
    elif (('10K' in text) or ('10,000' in text)):
        return 10000
    elif (('1K' in text) or ('1,000' in text) or('1000' in text)):
        return 1000
    else:
        return 0


dict_A = {}
for page_name, link in link_list.items():
    Test_page = 'https://worldpopulationreview.com'
    page = 'https://worldpopulationreview.com'
    degel = 0
    number = 0
    respones = requests.get(page + link)
    page = bs(respones.text, 'html.parser')
    time.sleep(5)
    table = page.find('tbody', {'class': 'jsx-2006211681'})
    if (table):
        pass  ## there is a Table in the href so we take the first
    else:
        continue  ## if there no table theres nothing to scrap in the href so we continue the loop
    if (page.find('div', {'class': 'note'}) ):
        number = func_find_number(str(page.find('div', {'class': 'note'}).text))
        if (number != 0):
            degel = 1
          
    if (degel == 0):
        if (page.find('tr', {'class': 'jsx-2006211681'})):
            for i in page.find_all('th'):
                degel = 0
                if (i.text):
                    number = func_find_number(str(i.text))
                    if (number != 0):
                        degel = 1
                        break
    print(page_name, Test_page + link    ,degel)
    for row in table.find_all('tr'):
        for tab in table.find_all('td'):
            if ((tab.text in help_country_arr) and (tab.text in dict_countrys)):
                if (degel == 1):
                    if ((tab.find_next("td").text)):
                        print(True)
                        value = (float((tab.find_next("td").text).replace(',', '')) * dict_countrys[tab.text]) / number
                        if (tab.text in dict_A):
                            dict_A[tab.text].update({page_name: value})
                        else:
                            dict_A[tab.text] = {page_name: value}
                    else:
                        if (tab.text in dict_A):
                            dict_A[tab.text].update({page_name: tab.find_next("td").text})
                        else:
                            dict_A[tab.text] = {page_name: tab.find_next("td").text}
                else:
                    if (tab.text in dict_A):
                        dict_A[tab.text].update({page_name: tab.find_next("td").text})
                    else:
                        dict_A[tab.text] = {page_name: tab.find_next("td").text}
            else:
                pass

    #print(dict_A)
```
